```python
def change(X):
    def label(x):
        if(x['label']=='normal'):
            return 0
        return 1
    def Proto(conf):
        if (not pd.isna(conf['arp.opcode']) ):
            return 'arp'
        elif (not pd.isna(conf['icmp.type']) ):
            return 'icmp'
        elif (not pd.isna(conf['ipv6.src']) ):
            return 'IPV6'
       # elif (~np.isnan(conf['SrcAddr']) ):
        return 'IPV4'
    #X['label']=X.apply(lambda x:label(x),axis = 1)
    X['Proto']=X.apply(lambda x:Proto(x),axis = 1)
    f=X

    srcports=[]
    dstports=[]
    srcips=[]
    dstips=[]
    for index,row in f.iterrows():
        IPtype = np.nan
        timestamp = row[0]
        framelen = row[1]
        srcIP = ''
        dstIP = ''
        if not pd.isna(row[4]):  # IPv4
            srcIP = row[4]
            dstIP = row[5]
            IPtype = 0
        elif not pd.isna(row[17]):  # ipv6
            srcIP = row[17]
            dstIP = row[18]
            IPtype = 1
        #print(row[6],row[8])
        srcproto = (row[6] if (not pd.isna(row[6])) else row[8])

        # UDP or TCP port: the concatenation of the two port strings will will results in an OR "[tcp|udp]"
        dstproto = (row[7] if (not pd.isna(row[7])) else row[9])
    #    print(row[7],row[9],dstproto)
    #     srcMAC = row[2]
    #     dstMAC = row[3]
        if pd.isna(srcproto):  # it's a L2/L1 level protocol
            if not pd.isna(row[12]):  # is ARP
                srcproto = 'arp'
                dstproto = 'arp'
                srcIP = row[14]  # src IP (ARP)
                dstIP = row[16]  # dst IP (ARP)
                IPtype = 0
            elif not pd.isna(row[10]):  # is ICMP
                srcproto = 'icmp'
                dstproto = 'icmp'
                IPtype = 0
#             elif srcIP + srcproto + dstIP + dstproto == '':  # some other protocol
#                 srcIP = row[2]  # src MAC
#                 dstIP = row[3]  # dst MAC 
    #     IPtypes.append(IPtype)
        srcips.append(srcIP)
        dstips.append(dstIP)
        srcports.append(srcproto)
        dstports.append(dstproto)
    X['srcports']=srcports
    X['dstports']=dstports
    #X['ip.src']=srcips
    #X['ip.dst']=dstips
```
